```python
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D
from keras.utils.vis_utils import plot_model
```

```python
def vgg_block(layer_in, n_filters, n_conv):
    for _ in range(n_conv):
        layer_in = Conv2D(n_filters, (3,3), activation='relu', padding='same')(layer_in)
    
    layer_in = MaxPooling2D((2,2), strides=(2,2))(layer_in)
    return layer_in
```

```python
visible = Input(shape=(256,256,3))
layer1 = vgg_block(visible, 64, 2)
layer2 = vgg_block(layer1, 128, 2)
layer3 = vgg_block(layer2, 256, 4)
model = Model(inputs=visible, outputs=layer3)
```

```python
model.summary()
```

```python
plot_model(model, show_shapes=True, to_file='multiple8_vgg_block.png')
```
